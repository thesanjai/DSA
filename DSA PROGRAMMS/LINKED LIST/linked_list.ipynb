```python
import random

class node:
    def __init__(self,data):
        self.data = data
        self.next = None

class linked_list:
    def __init__(self):
        self.first = self.last = None

    #Insertion at back
    def append(self,data):
        p = node(data)
        if self.first is None:
            self.first = self.last = p
            return
        self.last.next = p
        self.last = p

    #Insertion at front
    def appendleft(self,data):
        p = node(data)
        if self.first is None:
            self.first = self.last = p
            return
        p.next = self.first
        self.last = self.first
        self.first = p

    #displaying the linked list
    def display(self):
        if self.first is None:
            print("linked list is Empty")
            return
        p = self.first
        while p:
            print(p.data,end=" ")
            p = p.next

    #Reverse order usinf recursion
    def Rdisplay(self,head):
        if head:
            self.Rdisplay(head.next)
            print(head.data,end=' ')

    def display_rev(self):
        self.Rdisplay(self.first)

    #counting the nodes in linked list
    def len(self):
        if self.first is None:
            return 0
        count = 0
        p = self.first
        while p:
            count +=1
            p = p.next

        return count
        


def main():
    n = int(input('Enter the number of Nodes : ').strip())
    
    arr = []
    for i in range(n):
        t = random.randint(0,100)
        arr.append(t)

    print(f"Initial Array Values: {arr}")

    ll = linked_list()
    print("length of the ll is ",ll.len())

    for data in arr:
        ll.append(data)
        
    ll.display()
    print()
    print("length of the ll is ",ll.len())

    # ll.display_rev()
    for data in arr:
        ll.appendleft(data)
        
    ll.display()
    print()
    print("length of the ll is ",ll.len())


if __name__ == '__main__':
    main()
        
```
