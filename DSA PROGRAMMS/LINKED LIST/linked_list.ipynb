```python
import random

class node:
    def __init__(self,data):
        self.data = data
        self.next = None

class linked_list:
    def __init__(self):
        self.first = self.last = None

    #Insertion at back
    def append(self,data):
        p = node(data)
        if self.first is None:
            self.first = self.last = p
            return
        self.last.next = p
        self.last = p

    #Insertion at front
    def appendleft(self,data):
        p = node(data)
        if self.first is None:
            self.first = self.last = p
            return
        p.next = self.first
        self.last = self.first
        self.first = p

    #displaying the linked list
    def display(self,head=None):
        if head is None:
            if self.first is None:
                print("linked list is Empty")
                return
            p = self.first
            while p:
                print(p.data,end=" ")
                p = p.next
            return
        if head is None:
            print("linked list is Empty")
            return
        p = head
        while p:
            print(p.data,end=" ")
            p = p.next


    #Reverse order usinf recursion
    def Rdisplay(self,head):
        if head:
            self.Rdisplay(head.next)
            print(head.data,end=' ')

    def display_rev(self):
        self.Rdisplay(self.first)

    #counting the nodes in linked list
    def len(self):
        if self.first is None:
            return 0
        count = 0
        p = self.first
        while p:
            count +=1
            p = p.next

        return count
    
    # Reverse the linked list
    def reverse(self):
        if self.first is None:
            return
        r = q = None
        p = self.first
        while p:
            r = q
            q = p
            p = p.next
            q.next = r
       
        self.first = q

    #merrging two linked lists space complexity O(n)
    def merge(self, ll2):
        if self.first is None:
            return ll2
        if ll2.first is None:
            return self
        
        p1 = self.first
        p2 = ll2.first
        merged_list = linked_list()
        
        while p1 and p2:
            if p1.data < p2.data:
                merged_list.append(p1.data)
                p1 = p1.next
            else:
                merged_list.append(p2.data)
                p2 = p2.next
        
        while p1:
            merged_list.append(p1.data)
            p1 = p1.next
        
        while p2:
            merged_list.append(p2.data)
            p2 = p2.next
        
        return merged_list
    
    #merge two linked lists in place O(1) space complexity
    def merge_in_place(self, ll2):
        if self.first is None:
            return ll2
        if ll2.first is None:
            return self
        
        p1 = self.first
        p2 = ll2.first
        third = last = None
        if p1.data < p2.data:
            third = last = p1
            p1 = p1.next
            last.next = None
        else:
            third = last = p2
            p2 = p2.next
            last.next = None
        while p1 and p2:
            if p1.data < p2.data:
                
                last.next = p1
                last = p1
                p1 = p1.next
            else:
                last.next = p2
                last = p2
                p2 = p2.next
        if p1:
            last.next = p1
        if p2:
            last.next = p2
        return third      

def main():
    n = int(input('Enter the number of Nodes : ').strip())
    
    arr1 = []
    for i in range(n):
        t = random.randint(0,100)
        arr1.append(t)

    arr2 = []
    for i in range(n):
        t = random.randint(0,100)
        arr2.append(t)

    arr1.sort()
    arr2.sort()
    print(f"Initial Array Values 1: {arr1}")
    print(f"Initial Array Values 2: {arr2}")

    ll1  = linked_list()
    ll2  = linked_list()
    for data in arr1:
        ll1.append(data)
    for data in arr2:
        ll2.append(data)


    print("Linked List 1:")
    ll1.display()
    print()
    print("\nLinked List 2:")
    ll2.display()
    print()

    merged_ll = ll1.merge(ll2)
    print("\nMerged Linked List:")
    merged_ll.display()
    print()
    
    merged_ptr = ll1.merge_in_place(ll2)
    print("\nMerged Linked List In Place:")
    ll1.display(merged_ptr)
    print()
    

if __name__ == '__main__':
    main()
        
```
